```python
# Define the profiles to search
profiles = ['zanhansen', 'zulieena']  # Replace with the actual profile usernames

# Define the keywords to search for in captions
keywords = ['good', '2020', 'life']

# Define the maximum number of posts to download
max_posts = 10

# Function to download posts from specified profiles that contain certain keywords in their captions
def download_keyword_filtered_posts(profiles, keywords, max_posts):
    downloaded_count = 0

    for profile in profiles:
        try:
            # Get posts from the profile
            posts = instaloader.Profile.from_username(L.context, profile).get_posts()

            # Download each post that contains the keywords
            for post in posts:
                caption = post.caption.lower() if post.caption else ''
                if any(keyword.lower() in caption for keyword in keywords):
                    try:
                        print(f"Downloading post {post.shortcode} from profile {profile}")
                        L.download_post(post, target=f'{profile}_{post.date_utc.strftime("%Y-%m-%d")}')
                        downloaded_count += 1
                        if downloaded_count >= max_posts:
                            print(f"Downloaded {downloaded_count} posts. Stopping.")
                            return  # Exit function after reaching max posts
                        time.sleep(2)  # Pause to avoid rate limits
                    except instaloader.exceptions.InstaloaderException as e:
                        print(f"Error downloading post {post.shortcode}: {e}")
                        time.sleep(10)  # Longer pause on error
                else:
                    print(f"Skipping post {post.shortcode}, does not contain the keywords.")
        except instaloader.exceptions.InstaloaderException as e:
            print(f"An error occurred while processing profile {profile}: {e}")

# Download posts for the specified profiles and keywords
download_keyword_filtered_posts(profiles, keywords, max_posts)
```

```python
# Define the hashtag
hashtag = 'zulieena'

# Function to download posts for a given hashtag
def download_hashtag_posts(hashtag):
    try:
        # Get posts for the given hashtag
        posts = instaloader.Hashtag.from_name(L.context, hashtag).get_posts()

        # Download each post
        for post in posts:
            print(f"Downloading post {post.shortcode}")
            L.download_post(post, target=f'{hashtag}_{post.date_utc.strftime("%Y-%m-%d")}')
    except instaloader.exceptions.InstaloaderException as e:
        print(f"An error occurred: {e}")

# Download posts for the specified hashtag
download_hashtag_posts(hashtag)
```
